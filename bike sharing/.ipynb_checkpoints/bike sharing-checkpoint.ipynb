## Data Fields
- datetime - hourly date + timestamp  
- season 
    - 1 = spring
    - 2 = summer
    - 3 = fall
    - 4 = winter 
- holiday - whether the day is considered a holiday
- workingday - whether the day is neither a weekend nor holiday
- weather 
    - 1: Clear, Few clouds, Partly cloudy, Partly cloudy
    - 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
    - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
    - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog 
- temp - temperature in Celsius
- atemp - "feels like" temperature in Celsius
- humidity - relative humidity
- windspeed - wind speed
- casual - number of non-registered user rentals initiated
- registered - number of registered user rentals initiated
- count - number of total rentals

## Evaluation
Submissions are evaluated one the Root Mean Squared Logarithmic Error .

분류와 회귀중 회귀와 관련된 문제 => 자전거 대여량을 예측하는 문제이기 때문에

머신러닝의 학습 방법
1. 지도학습(Supervised Learning) - 정답을 알려주며 학습시키는 것
    - 분류(classification)
        - 이진 분류: 어떤 데이터에 대해 두 가지 중 하나로 분류할 수 있는 것
        - 다중 분류: 어떤 데이터에 대해 여러 값 중 하나로 분류할 수 있는 것
    - 회귀(regression)
        - 어떤 데이터들의 특징(feature)을 토대로 값을 예측하는 것
2. 비지도학습(Unsupervised Learning)
    - 정답을 따로 알려주지 않고(label이 없다), 비슷한 데이터들을 군집화 하는 것
3. 강화학습(Reinforcement Learning)
    - 상과 벌이라는 보상(reward)을 주며 상을 최대화하고 벌을 최소화 하도록 강화 학습하는 방식

### HeatMap
- A Heat Map shows the relationship between two items and provides rating information, such as high to low or poor to excellent.

```python
# corr = correlation 상관관계!
# 인덱스와 칼럼의 상관계수가 1에 가까울수록, 하나가 증가할 때 다른 하나도 함께 증가하는 경향이 있음
# 인덱스와 칼럼의 상관계수가 -1에 가까울수록, 하나가 증가할 때 다른 하나는 감소하는 경향이 있음
corrMatt = train[['temp', 'atemp', 'casual', 'registered', 'humidity', 'windspeed', 'count']]
corrMatt = corrMatt.corr()
print(corrMatt)

# 삼각형 마스크를 만들어서,
mask = np.array(corrMatt)
# 위쪽 삼각형을 False로 만들어 주어
mask[np.tril_indices_from(mask)] = False
# 반대로 하려면!
# mask[np.triu_indices_from(mask)] = False
# 더 보기 좋게 만든다!
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
# vamx는 Values to anchor the colormap, 
# otherwise they are inferred from the data and other keyword arguments.
# square는 True로 줘야 예쁘게 잘 보임
# annot는 True로 줘야 수치가 안에 들어가서 보임
sns.heatmap(corrMatt, mask=mask, vmax=.8, square=True, annot=True)
```

- 온도(temp), 습도(humidity), 풍속(windspeed)은 거의 연관관계가 없다.
- 대여량과 가장 연관이 높은 건 registered로 등록된 대여자가 많지만, test 데이터에는 이 값이 없다.
- atempd와 temp는 0.98로 상관관계가 높지만 온도와 체감온도로 피쳐로 사용하기에 적합하지 않을 수 있다.

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3)
fig.set_size_inches(12, 5)

# scatterplot은 값을 x축, y축에 맞게 그려주는 그래프
# regplot은 scatterplot + 회귀선을 그려주는 그래프
sns.regplot(data=train, x='temp', y='count', ax=ax1)
sns.regplot(data=train, x='windspeed', y='count', ax=ax2)
sns.regplot(data=train, x='humidity', y='count', ax=ax3)
```

```python
def concatenate_year_month(datetime):
    return "{0}-{1}".format(datetime.year, datetime.month)

train['year_month'] = train['datetime'].apply(concatenate_year_month)

print(train.shape)
# 특정 데이터만 보고 싶을 때 이런 식으로 볼 수도 있구나...
train[['datetime', 'year_month']].head()
```

```python
# trainWithoutOutliers
# 이상치 데이터 제거
# pandas.Series.std(표준편차): Return sample standard deviation over requested axis.
'''
편차(deviation)는 관측값에서 평균 또는 중앙값을 뺀 것이다.
분산(variance)은 관측값에서 평균을 뺀 값을 제곱하고, 
그것을 모두 더한 후 전체 개수로 나눠서 구한다. 
즉, 차이값의 제곱의 평균이다. 
관측값에서 평균을 뺀 값인 편차를 모두 더하면 0이 나오므로 제곱해서 더한다.
표준 편차(standard deviation)는 분산을 제곱근(루트)한 것이다.
'''
# 이걸 왜 이상치라고 판단하는 기준이 되었을까???
trainWithoutOutliers = train[np.abs(train["count"] - train["count"].mean()) <= (3*train["count"].std())]
```
